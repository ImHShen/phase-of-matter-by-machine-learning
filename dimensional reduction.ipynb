```python
%matplotlib inline
from __future__ import division
import numpy as np
from numpy.random import rand
from sklearn.decomposition import PCA
from sklearn import manifold
import matplotlib.pyplot as plt
```

```python
def config(n_potts=3):
    file_name=''.join(["config_table_add",str(n_potts),".txt"])
    data=np.loadtxt(file_name)
    configurations=data[:,:-1]
    temperature=data[:,-1]
    config_x=np.cos(2*np.pi*configurations/n_potts)
    config_y=np.sin(2*np.pi*configurations/n_potts)
    config_expand=np.append(config_x,config_y,axis=1)
    return [config_expand,temperature]
```

```python
config2,temperature2=config(n_potts=2)
pca = PCA(n_components=2)
projected_config=pca.fit_transform(config2)
plt.scatter(projected_config[:,0],projected_config[:,1],c=temperature2)
plt.colorbar()
```

```python
config3,temperature3=config(n_potts=3)
pca = PCA(n_components=2)
projected_config=pca.fit_transform(config3)
plt.scatter(projected_config[:,0],projected_config[:,1],c=temperature3)
plt.colorbar()
```

```python
tsne=manifold.TSNE(n_components=2)
config_tsne=tsne.fit_transform(config2)
plt.scatter(config_tsne[:,0],config_tsne[:,1],c=temperature2)
```
