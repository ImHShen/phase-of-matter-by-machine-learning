```python
%matplotlib inline
from __future__ import division
import numpy as np
from numpy.random import rand
from sklearn.decomposition import PCA
from sklearn import manifold
import matplotlib.pyplot as plt
```

```python
L=32
data=np.loadtxt("data/output_config.dat")
len_s=data.shape[0]
temperature=data[:,-1]
configurations=data[:,:-1].reshape((len_s,L*L))
configurations=(configurations+np.pi)/(2*np.pi)
config_x=np.cos(configurations)
config_y=np.sin(configurations)
config_expand=np.append(config_x,config_y,axis=1)
```

```python
pca = PCA(n_components=2)
projected_config=pca.fit_transform(config_expand)
pic= plt.scatter(projected_config[:,0],projected_config[:,1],c=temperature)
```

# Since KT phase transition is marked by different behavior of spin correlation function, we first generate correlation data and then apply PCA.

```python
def autocorrelation(i,j,config2d):
    nx=config2d.shape[0]
    ny=config2d.shape[1]
    sums=0
    for ix in range(nx):
        for jx in range(ny):
            sums+=config2d[ix,jx]*config2d[(ix+i)%nx,(jx+j)%ny]
    return(sums/(nx*ny))            
```

```python
def autocorrelation_config(configs,nx,ny):
    n_sample=configs.shape[0]
    new_configs=[]
#    new_configs=np.array(new_configs)
    for n in range(n_sample):       
        config2d=configs[n,:].reshape((nx,ny))
        corr=np.zeros((nx,ny))
        for i in range(nx):
            for j in range(ny):
                corr[i,j]=autocorrelation(i,j,config2d)
        new_configs.extend(corr.reshape((1,nx*ny)))
#        np.append(new_configs,corr.reshape((1,nx*ny)),axis=0)
    return(np.array(new_configs))        
```

```python
corr_x=autocorrelation_config(config_x,L,L)
corr_y=autocorrelation_config(config_y,L,L)
corr_expand=np.append(corr_x,corr_y,axis=1)
```

```python
pca = PCA(n_components=2)
n_show=corr_expand.shape[0]
projected_config=pca.fit_transform(corr_expand)
plt.scatter(projected_config[:n_show,0],projected_config[:n_show,1],c=temperature[:n_show])
plt.colorbar()
plt.show()
```
